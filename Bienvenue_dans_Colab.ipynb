</div>

Colab vous permet de tirer pleinement parti des biblioth√®ques populaires Python pour analyser et visualiser des donn√©es. La cellule de code ci-dessous utilise <strong>numpy</strong> pour g√©n√©rer des donn√©es al√©atoires et <strong>matplotlib</strong> pour les visualiser. Pour modifier le code, cliquez simplement sur la cellule.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

### Distributions des Variables Num√©riques Cl√©s

```
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Distributions des Variables Num√©riques Cl√©s', fontsize=16)

sns.histplot(df['Age'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution de l\'√Çge')

sns.histplot(df['Daily_Screen_Time(hrs)'], kde=True, ax=axes[0, 1], color='lightcoral')
axes[0, 1].set_title('Distribution du Temps d\'√âcran Quotidien (heures)')

sns.histplot(df['Sleep_Quality(1-10)'], kde=True, ax=axes[0, 2], color='lightgreen')
axes[0, 2].set_title('Distribution de la Qualit√© du Sommeil (1-10)')

sns.histplot(df['Stress_Level(1-10)'], kde=True, ax=axes[1, 0], color='gold')
axes[1, 0].set_title('Distribution du Niveau de Stress (1-10)')

sns.histplot(df['Days_Without_Social_Media'], kde=True, ax=axes[1, 1], color='mediumpurple')
axes[1, 1].set_title('Distribution des Jours sans M√©dias Sociaux')

sns.histplot(df['Happiness_Index(1-10)'], kde=True, ax=axes[1, 2], color='salmon')
axes[1, 2].set_title('Distribution de l\'Indice de Bonheur (1-10)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Distributions des Variables Cat√©gorielles Cl√©s

```
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Distributions des Variables Cat√©gorielles Cl√©s', fontsize=16)

sns.countplot(y=df['Gender'], ax=axes[0], palette='viridis')
axes[0].set_title('Distribution du Genre')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Gender')

sns.countplot(y=df['Social_Media_Platform'], ax=axes[1], palette='magma')
axes[1].set_title('Distribution des Plateformes de M√©dias Sociaux')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Social Media Platform')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Relations entre le Temps d'√âcran, le Niveau de Stress et l'Indice de Bonheur

```
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Daily_Screen_Time(hrs)', y='Happiness_Index(1-10)', hue='Stress_Level(1-10)', size='Stress_Level(1-10)', palette='coolwarm', data=df)
plt.title('Temps d\'√âcran vs. Indice de Bonheur avec Niveau de Stress')
plt.xlabel('Temps d\'√âcran Quotidien (heures)')
plt.ylabel('Indice de Bonheur (1-10)')
plt.show()
```

### Niveau de Stress par Qualit√© de Sommeil

```
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep_Quality(1-10)', y='Stress_Level(1-10)', palette='YlGnBu', data=df)
plt.title('Niveau de Stress par Qualit√© de Sommeil')
plt.xlabel('Qualit√© du Sommeil (1-10)')
plt.ylabel('Niveau de Stress (1-10)')
plt.show()
```
